# 02 - EDA (Univariate + Bivariate)

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from src.utils.metrics import target_rate_by_group

TRAIN_PATH = "data/raw/train.csv"
TARGET = "target"
df = pd.read_csv(TRAIN_PATH)
```

```python

num_cols = df.select_dtypes(include="number").columns.drop(TARGET, errors="ignore")
cat_cols = df.select_dtypes(exclude="number").columns

df[num_cols].describe().T.head(20)
```

```python

col = num_cols[0]
fig, ax = plt.subplots(1,2, figsize=(12,4))
sns.histplot(df[col], ax=ax[0], kde=True)
sns.boxplot(x=df[col], ax=ax[1])
plt.show()
```

```python

if len(cat_cols) > 0:
    gcol = cat_cols[0]
    target_rate_by_group(df, gcol, TARGET).head(20)
```
